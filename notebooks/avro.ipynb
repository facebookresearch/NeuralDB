```python
import pprint
from avro.datafile import DataFileReader
from avro.io import DatumReader

schema = "../TripleLinked.avsc"
data = "../OPIEC-Clean-example.avro"
pp = pprint.PrettyPrinter()
reader = DataFileReader(open(data, "rb"), DatumReader())

```

```python
def get_tokens(item):
    return " " .join([tok["word"] for tok in item])


for i, article in enumerate(reader):
    #print(article.keys())
    #print(article["triple_id"])

    print([tok['word'] for tok in article["sentence_linked"]['tokens']])
    print(" ".join([get_tokens(article["subject"]), get_tokens(article["relation"]), get_tokens(article["object"])]))
    print()
    #pp.pprint(article["subject"])
    #print(article["sentence_linked"])
    #    pp.pprint(article)

    if i > 100:
        break
```
