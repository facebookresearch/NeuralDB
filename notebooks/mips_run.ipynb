```python
mips_types = ["t5","bert"]

def get_cnn_sizes():
    return [16]

def get_models():
    return [["t5", ["CNN",get_cnn_sizes]], ["bert"]]

def get_beta():
    return [0.1, 0.5, 1.0, 2, 5, 10]

def n_layers_bias():
    return [1,2,4,6,8,12]

def get_lr():
    return [1e-2, 8e-3, 5e-3, 2e-3, 1e-3, 8e-4, 5e-4, 2e-4, 1e-4]

def grad_steps():
    return [4,8,16,32,64]
```

```python
import itertools

all_expts = list(itertools.product(grad_steps(), get_lr(), n_layers_bias(), get_beta(), get_models()))
context = 50
version="v0.2"

print(len(all_expts))

jobs = []
for grad, lr, layers, beta, (model,*args) in all_expts:
    base = "bash -x scripts/train.t5.rec.sh mips_sweep {} {} {} {} {} {} {}".format(model, lr, version, context, beta, layers, grad )

    if len(args):
        extra_args = []
        for k,v in args:
            for choice in v():
                jobs.append("{}={} {}".format(k,choice,base))
    else:
        jobs.append(base)


print(len(jobs))


import random

random.shuffle(jobs)
```

```python
with open("sweep_mips.sh","w+") as f:
    for job in jobs:
        f.write(job+"\n")
        print(job)

```
