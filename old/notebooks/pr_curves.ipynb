```python
import matplotlib.pyplot as plt

a = [[0.732514814, 0.9989402412],
[0.8396888158,0.9950792657],
[0.9010741479,0.9777747208],
[0.9484177416,0.8724650406],
[0.9674129884,0.7018277109],
[0.980976907, 0.6359623463]]

b = [[0.3129041032,0.9950748777],
[0.5247405783,0.9872845573],
[0.720254584,0.9584454733],
[0.8591060554,0.8568413421],
[0.9630613563,0.6183004355],
[0.9939527213,0.3231524589]]

a = [[c*100,d*100] for d,c in a]
b = [[c*100,d*100] for d,c in b]
```

```python
with plt.style.context("ggplot"):
    plt.figure(figsize=(6,4))
    p,r = list(zip(*a))
    #plt.scatter(p,r)
    plt.plot(p,r,"-*")

    p,r = list(zip(*b))
    #plt.scatter(p,r,"+")
    plt.plot(p,r,"-o")

    plt.xlabel("Recall (%)")
    plt.ylabel("Precision (%)")
    plt.xlim((50,100))
    plt.ylim((50,100))
    plt.title("Precision/Recall Trade-off for SSG")

    plt.legend(["Full Spervision", "Distant Supervision"])
    plt.savefig("/scratch/jth/pr.pdf",bbox_inches = 'tight')
```
