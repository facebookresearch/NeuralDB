```python
def get_dataset():
    return ["v0.2","v0.4"]

def get_context():
    return [100]

def get_filter_size():
    return [1,2,4,6,8,12,16,24,32]

def get_lr():
    return [2e-3, 1e-3, 8e-4, 5e-4, 2e-4, 1e-4, 8e-5, 6e-5]

```

```python
import itertools
all_expts = list(itertools.product(get_dataset(), get_context(), get_lr(), get_filter_size()))

print(len(all_expts))

jobs = []
for version, context, lr, filters in all_expts:
    base = "bash -x scripts/train.t5.fusion.sh tfidf_fusion {} {} {} {}".format(version, context, lr, filters )
    jobs.append(base)


print(len(jobs))


import random

random.shuffle(jobs)
```

```python
with open("sweep_fusion.sh","w+") as f:
    for job in jobs:
        f.write(job+"\n")
        print(job)

```
